```python
import pandas as pd
import numpy as np
import ggseg
```

## Displaying residuals using ggseg
- python-ggseg GitHub page: https://github.com/ggseg/python-ggseg/tree/main 

```python
# Dummy residuals
res_test = np.random.uniform(low=-0.5, high=0.5, size=68)
```

```python
# Prepare data for ggseg

# Get cortical region list
CORT_IDX = np.concatenate([np.arange(34), np.arange(49, 83)])
region_list = pd.read_csv( "../data/TauRegionList_ggseg.csv")["Raj_label_ggseg"].tolist()  
## naming convention for ggseg (https://github.com/ggseg/python-ggseg/tree/main/ggseg/data/dk)
ctx_region_list = list(map(lambda x: region_list[x],CORT_IDX))

# Create dictionary containing cortical regions and corresponding residuals
res_d = dict(zip(ctx_region_list, res_test))
```

```python
res_d
```

```python
# Display residuals
## N.B. frontalpole and temporalpole regions not shown

ggseg.plot_dk(res_d, cmap='RdBu',
              background='black', 
              edgecolor='grey', bordercolor='w',
              vminmax = [-0.5, 0.5],   
              figsize=(10,5),
              ylabel='model residuals', 
              title=f"ggseg test",
              )
```
